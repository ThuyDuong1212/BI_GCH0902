```python
#Process file manually 
f = open('authors.txt', 'r')

author_lines = f.readlines()

print('{0:20s}{1:20s}'.format('Name','Email'))
#Skip 3 row
author_lines = author_lines[3:]
for line in author_lines:
    items = line.split('<')
    name = items[0][:-1]
    email = items[-1][:-2]
    print('{0:30s}{1:20s}'.format(name,email))

f.close()
```

```python
# read to columns
IDs, names,stdgrades = np.loadtxt('gradebook.txt', dtype=str, skiprows=7, delimiter='\t',
                                    unpack=True, usecols=(2,3,4))
print('{0:20s}{1:20s}{2:5s}'.format('ID','Name','Grade'))

# print student name, ID grade of padded students(grade > 4)
for i in range(len(IDs)):
    if float (stdgrades[i]) >=4.0:
        print('{0:10s}{1:20s}{2:5s}'.format(IDs[i],names[i], stdgrades[i]))
```

```python
import csv

grades = []

f = open('gradebook.csv', 'r')
for row in csv.reader(f, delimiter=','):
    grades.append(row)

print(grades)
grades = grades[1:]
passed_students = [row for row in grades if float(row[-1]) >= 4.5]

print('Passed students')
print('{0:10s}{1:20s}{2:5s}'.format('ID', 'Name', 'Grade'))
for row in passed_students:
    print('{0:10s}{1:20s}{2:>5s}'.format(row[2], row[3], row[4]))
f.close()
```
